```python
import pandas as pd 
weather = pd.read_csv("weather.csv", index_col = "DATE") #importing the csv file
weather
```

```python
null_pct = weather.apply(pd.isnull).sum()/(weather).shape[0] # find null values in the dataset
null_pct
```

```python
valid_columns = weather.columns[null_pct < .05] # remove columns with no value larger than 5%
valid_columns
```

```python
weather = weather[valid_columns].copy() # copy to avoid SettingWithCopyWarning
weather.columns = weather.columns.str.lower()
weather
```

```python
weather = weather.ffill() #fill missing values
```

```python
weather["target"] = weather.shift(-1)["tmax"] # create a target variable
weather
```

```python
weather = weather.ffill() # fill the target variable
weather
```

```python
from sklearn.linear_model import Ridge # import Ridge model
rr = Ridge(alpha = 0.1) # create a Ridge model with alpha = 0.1(controls how much coef are shrunk)
```

```python
predictors = weather.columns[~weather.columns.isin(["target", "name", "station"])] # create predictors variable except target, name and station
predictors
```

```python
def backtest(weather, model, predictors, start=3650, step=90): # function to backtest the model
    all_predictions = []
    
    for i in range(start, weather.shape[0], step): # loop through the dataset with a step size of 90 days
        train = weather.iloc[:i,:] # training data
        test = weather.iloc[i:(i+step),:]# testing data
        
        model.fit(train[predictors], train["target"])# fit the model on the training data
        # make predictions on the test data
        
        preds = model.predict(test[predictors])# predict the target variable
        preds = pd.Series(preds, index=test.index)# create a series with the index of the test data
        combined = pd.concat([test["target"], preds], axis=1)# combine the target and prediction data
        combined.columns = ["actual", "prediction"]# rename the columns
        combined["diff"] = (combined["prediction"] - combined["actual"]).abs()# calculate the absolute difference between the prediction and actual values
        
        all_predictions.append(combined)# append the combined data to the list
    return pd.concat(all_predictions)
```

```python
predictions = backtest(weather, rr, predictors) # call the backtest function)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error # import mean absolute error and mean squared error

mean_absolute_error(predictions["actual"], predictions["prediction"])
```

```python
def pct_diff(old, new):
    return (new - old) / old * 100 # create a function to calculate the percentage difference

def compute_rolling(weather,horizon, col):
    label = f"rolling_{horizon}_{col}"
    
    weather[label] = weather[col].rolling(horizon).mean() # create a rolling mean
    weather[f"{label}_pct_diff"] = pct_diff(weather[label], weather[col]) # create a percentage difference column
    return weather # return the weather data with the new columns

rolling_horizons = [3, 14] # create a list of rolling horizons

for horizon in rolling_horizons: # loop through the rolling horizons
    for col in ["tmax", "tmin", "prcp"]:
        weather = compute_rolling(weather, horizon, col) # call the compute rolling function

weather
```

```python
weather = weather.iloc[14:,:]# remove the first 14 rows to avoid NaN values
weather = weather.fillna(0) # fill NaN values with 0
weather
```

```python
def expand_mean(df):
    return df.expanding(1).mean()

for col in ["tmax", "tmin", "prcp"]:
    weather[f"expanding_mean_{col}"] = (weather[col]).groupby(weather.index.month, group_keys = False).apply(expand_mean) # create a expanding mean column
    weather[f"day_avg_{col}"] = weather[col].groupby(weather.index.day_of_year, group_keys=False).apply(expand_mean) # create a day average column
```
